```python
import pandas as pd
import numpy as np
import math as m
```

```python
data = pd.read_csv('tennis.csv')
df = pd.DataFrame(data)
df
```

```python
from sklearn import preprocessing
import numpy as np
from sklearn.preprocessing import MinMaxScaler
# ordinal encode attributes
enc = preprocessing.OrdinalEncoder()
enc.fit(df)
X = enc.transform(df)
X1 = X[:,0:4]

Y=X[:,4]
print(X1, Y)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy',random_state=0, max_depth=3, min_samples_leaf=2)
clf = model.fit(X1,Y) #train your classifier using preprocessed data, saying attributes (X) and label (Y)
```

```python
fn=['Outlook','Temperature','Humidity','Wind']
cn=['no', 'yes']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('TennisEntropy.png')
```

```python
print(Y)
Y1 = clf.predict(X1)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
print(accuracy_score(Y, Y1)) 
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y, Y1)
```

```python
from sklearn import metrics
print(metrics.precision_score(Y, Y1))
print(metrics.recall_score(Y, Y1))
print(metrics.f1_score(Y, Y1))
```

```python
import numpy as np
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(Y, Y1, pos_label=2)
print(fpr)
print(tpr)
print(thresholds)
auc_score = roc_auc_score(Y, Y1)
print(auc_score)
from matplotlib import pyplot as plt
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='ROC')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
```
