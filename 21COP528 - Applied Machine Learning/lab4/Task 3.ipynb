```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import metrics
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
%matplotlib inline
X1, y1=datasets.make_circles(n_samples=5000, factor=.6,
 noise=.05)
X2, y2 = datasets.make_blobs(n_samples=1000, n_features=2, centers=[[1.2,1.2]],
cluster_std=[[.1]],
 random_state=9)
X = np.concatenate((X1, X2))
plt.scatter(X[:, 0], X[:, 1], marker='o')
plt.show()

```

```python
from sklearn.cluster import DBSCAN
y_pred = DBSCAN().fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
y_pred = DBSCAN(eps = 0.1, min_samples = 10).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
cluster_KMean = KMeans(n_clusters=5,random_state=123).fit(X)
y_pred_KMean = cluster_KMean.labels_
s_Score_KMean=metrics.silhouette_score(X, y_pred_KMean, metric='euclidean', sample_size=None,
random_state=None)
# Visualising
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred_KMean)
plt.title('KMean')
plt.text(.95, .05, ('n_clusters=%d, s_score: %.2f' % (3,s_Score_KMean)),
 transform=plt.gca().transAxes, size=10,
 horizontalalignment='right')
plt.show()
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/lborohfang/COP528AML_Files/main/faithful_data.csv')
print(df.head(10))
faithful_data = np.asarray(df.iloc[:,0:])
faithful_data[:,0]
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram , linkage

```

```python
Z = linkage(faithful_data, method = 'average')
plt.figure(figsize=(10,6))
#plotting dendrogram
dendro = dendrogram(Z)
plt.title('Dendrogram')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
plt.scatter(faithful_data[:,0], faithful_data[:,1], marker='o')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
y_pred = DBSCAN(eps = 2, min_samples = 3).fit_predict(faithful_data)


plt.show()
from sklearn import metrics
silhouetteScore=metrics.silhouette_score(faithful_data, y_pred, metric='euclidean', sample_size=None,
random_state=None)

plt.scatter(faithful_data[:,0], faithful_data[:,1], c=y_pred)
plt.title('DBSCAN')
plt.text(.95, .05, ('eps=%d, s_score: %.2f' % (3,silhouetteScore)),
 transform=plt.gca().transAxes, size=10,
 horizontalalignment='right')
```

```python
cluster_KMean = KMeans(n_clusters=2,random_state=123).fit(faithful_data)
y_pred_KMean = cluster_KMean.labels_
s_Score_KMean=metrics.silhouette_score(faithful_data, y_pred_KMean, metric='euclidean', sample_size=None,
random_state=None)
# Visualising
plt.figure(figsize=(10, 6))
plt.scatter(faithful_data[:, 0], faithful_data[:, 1], c=y_pred_KMean)
plt.title('KMean')
plt.text(.95, .05, ('n_clusters=%d, s_score: %.2f' % (2,s_Score_KMean)),
 transform=plt.gca().transAxes, size=10,
 horizontalalignment='right')
plt.show()
```
