```python
from sklearn import datasets
df=datasets.make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0,
n_repeated=0,n_classes=4, n_clusters_per_class=1, weights=None, flip_y=0.01, class_sep=1.0, hypercube=True,
shift=0.0, scale=1.0,shuffle=True, random_state=None)
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram , linkage
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df1 = df[0]

df1
df






```

```python
#Creating the model
agg_clustering = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'ward')
#predicting the labels
labels = agg_clustering.fit_predict(df1)
plt.scatter(df1[:, 0], df1[:, 1], c=labels, s=30, cmap='viridis')

from sklearn import metrics
silhouetteScore=metrics.silhouette_score(df1, labels, metric='euclidean', sample_size=None, random_state=None)
print(silhouetteScore)
```

```python
#Linkage Matrix
Z = linkage(df1, method = 'average')
plt.figure(figsize=(10,6))
#plotting dendrogram
dendro = dendrogram(Z)
plt.title('Dendrogram')
plt.ylabel('Euclidean distance')
plt.show()
```
